## Objetivos
- Deteção e tratamento de outliers com IQR
- Estrategias de mitiogação de valores extremos
- Codificação de variaveis categoricas
- Preparação final dos dados para a fase de modelação


Método IQR




```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df=pd.DataFrame({"idade":[20,25,30,35],
                "salario":[2000,2500,3000,4000]})

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
print(df_scaled)
```
